```python
import spacy
from spacy import displacy
import csv
nlp = spacy.load('en_core_web_sm')
import re
from pyjarowinkler import distance
```

```python
f = open('/home/a.mahalik/hafoor/test_set.txt', 'r')

data = f.read()
array1 = []
splat = data.split("~")
for number, paragraph in enumerate(splat, 1):
    array1 += [paragraph]
array1= array1[3:]
```

```python
list_company = []
id_company=[]
with open('/home/a.mahalik/hafoor/company_map.txt') as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content] 
for i in content:
    t = i.split(',')
    print(t)
    list_company.append(t[0])
    id_company.append(t[2])
list_company=list_company[1:]
id_company=id_company[1:]
for i in range(len(id_company)):
    id_company[i]=int(id_company[i])
```

```python
i=0
with open('final_test.csv', 'a') as csvFile:
    while i<len(array1):
        t = array1[i]+array1[i+1]
        i=i+2
        doc = nlp(t)
        organisation = []
        row = [int(i/2)]
        remove_able_temp = []
        for ent in doc.ents:
            if ent.label_ == 'ORG':
                organisation.append(ent.text)
        ans = ""
        for iq in range(len(organisation)):
            temp = organisation[iq]
            score = 0
            t_temp = 0
            for k in range(len(list_company)):
                score_temp = distance.get_jaro_distance(list_company[k], temp, winkler=True, scaling=0.1)
                if score_temp>score:
                    score = score_temp
                    t_temp = k
            yolo = str(id_company[t_temp])
            if (score>.90 )and (yolo not in remove_able_temp):
                remove_able_temp.append(yolo)
                # print(list_company[t_temp],organisation[iq])
        
        tatta = ' '.join(remove_able_temp)
        
        row.append(tatta)
        print(row)

        writer = csv.writer(csvFile)
        writer.writerow(row)
csvFile.close()
```
